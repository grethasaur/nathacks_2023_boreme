```python
# CHANNEL 1
# ============================================

import numpy as np
import matplotlib.pyplot as plt

# Assuming nb_ch1_ contains your channel data
nb_ch1_fft = np.fft.fft(nb_ch1)
nb_ch1_freq = np.fft.fftfreq(len(nb_ch1)) * 256

# Compute magnitude spectrum
magnitude_spectrum = np.abs(nb_ch1_fft[:len(nb_ch1) // 2])

# Plot the magnitude spectrum
plt.plot(nb_ch1_freq[:len(nb_ch1) // 2], magnitude_spectrum)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title('Channel 1 No Filter')
plt.show()

# Assuming nb_ch1_ contains your channel data
cut_freq = 60
tolerance = 2

# Apply Fourier Transform
nb_ch1_fft = np.fft.fft(nb_ch1)
nb_ch1_freq = np.fft.fftfreq(len(nb_ch1)) * 256

# Define a threshold frequency
threshold_freq = 100

# Remove frequencies greater than the threshold
nb_ch1_fft[np.abs(nb_ch1_freq) > threshold_freq] = 0


# Remove powerline interference
index_to_remove = np.where((nb_ch1_freq >= cut_freq - tolerance) & (nb_ch1_freq <= cut_freq + tolerance))
nb_ch1_fft[index_to_remove] = 0

# Plot the modified result
nb_ch1_plot_fft = nb_ch1_fft[:len(nb_ch1) // 2]
nb_ch1_plot_freq = nb_ch1_freq[:len(nb_ch1) // 2]

plt.plot(nb_ch1_plot_freq, np.abs(nb_ch1_plot_fft))  # Plot magnitude spectrum
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title('Channel 1 WITH Filter')
plt.show()

# PSD calculation after transform and filters
magnitude_spectrum = np.abs(nb_ch1_fft[:len(nb_ch1) // 2])
psd = (magnitude_spectrum**2) / (len(nb_ch1) * 256)

#Plot PSD
plt.plot(nb_ch1_freq[:len(nb_ch1) // 2], psd)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power Spectral Density (Î¼VÂ²/Hz)")
plt.title("Channel 1 Power Spectral Density")
plt.show()

# binning function
def computeBands(freq_array, filtered_data):
    bandTotals = [0, 0, 0, 0, 0]
    bandCounts = [0, 0, 0, 0, 0]

    for point in range(len(freq_array)):
        magnitude = np.abs(filtered_data[point])
        if freq_array[point] < 4 and magnitude > 0:
            bandTotals[0] += magnitude
            bandCounts[0] += 1
        elif freq_array[point] < 8 and magnitude > 0:
            bandTotals[1] += magnitude
            bandCounts[1] += 1
        elif freq_array[point] < 12 and magnitude > 0:
            bandTotals[2] += magnitude
            bandCounts[2] += 1
        elif freq_array[point] < 30 and magnitude > 0:
            bandTotals[3] += magnitude
            bandCounts[3] += 1
        elif freq_array[point] < 100 and magnitude > 0:
            bandTotals[4] += magnitude
            bandCounts[4] += 1

    # Calculate the average of all points
    bandAverages = np.array(bandTotals) / np.array(bandCounts)
    return bandAverages

# plot bins
def plotBands(bands):
    binNames = ["Delta", "Theta", "Alpha", "Beta", "Gamma"]
    plt.ylabel("Amplitude")
    plt.bar(binNames, bands, color="#7967e1")
    plt.title('Not Bored Band Distribution')
    plt.show()
    plt.clf()


# Example usage for Channel 1
nb_ch1_bands_before_ifft = computeBands(nb_ch1_freq, nb_ch1_fft)

# Plot bands before IFFT
plotBands(nb_ch1_bands_before_ifft)



# Step 3: Apply the Inverse Fourier Transform
# ============================================

# Create a copy of the Fourier-transformed data
nb_ch1_fft_copy = np.copy(nb_ch1_fft)

# Apply the Inverse Fourier Transform to the copy
nb_ch1_inverse_fft = np.abs(np.fft.ifft(nb_ch1_fft_copy))

time = np.arange(0, 256, 1)

plt.figure(figsize=(16, 10))
print("Filtered Plot")
plt.plot(time, nb_ch1_inverse_fft[:256])
plt.xlabel("Sample")
plt.ylabel("Î¼V")
plt.title('Channel 1 Inverse after filter')
plt.show()
plt.clf()
```

```python
b_ch1 = []
with open("bored.csv", "r") as data:
    for line in data:
        # Format the line
        line = line.split(",")         # Convert to a list
        b_ch1.append(float(line[0]))   # Just record the 1st channel

b_ch2 = []
with open("bored.csv", "r") as data:
    for line in data:
        # Format the line
        line = line.split(",")         # Convert to a list
        b_ch2.append(float(line[1]))   # Just record the 1st channel

b_ch3 = []
with open("bored.csv", "r") as data:
    for line in data:
        # Format the line
        line = line.split(",")         # Convert to a list
        b_ch3.append(float(line[2]))   # Just record the 1st channel

b_ch4 = []
with open("bored.csv", "r") as data:
    for line in data:
        # Format the line
        line = line.split(",")         # Convert to a list
        b_ch4.append(float(line[3]))   # Just record the 1st channel


# In[3]:


# CHANNEL 1
# ============================================

import numpy as np
import matplotlib.pyplot as plt

# Assuming b_ch1_ contains your channel data
b_ch1_fft = np.fft.fft(b_ch1)
b_ch1_freq = np.fft.fftfreq(len(b_ch1)) * 256

# Compute magnitude spectrum
magnitude_spectrum = np.abs(b_ch1_fft[:len(b_ch1) // 2])

# Plot the magnitude spectrum
plt.plot(b_ch1_freq[:len(b_ch1) // 2], magnitude_spectrum)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title('Channel 1 No Filter')
plt.show()

# Assuming b_ch1_ contains your channel data
cut_freq = 60
tolerance = 2

# Apply Fourier Transform
b_ch1_fft = np.fft.fft(b_ch1)
b_ch1_freq = np.fft.fftfreq(len(b_ch1)) * 256

# Define a threshold frequency
threshold_freq = 100

# Remove frequencies greater than the threshold
b_ch1_fft[np.abs(b_ch1_freq) > threshold_freq] = 0

# Remove powerline interference
index_to_remove = np.where((b_ch1_freq >= cut_freq - tolerance) & (b_ch1_freq <= cut_freq + tolerance))
b_ch1_fft[index_to_remove] = 0

# Plot the modified result
b_ch1_plot_fft = b_ch1_fft[:len(b_ch1) // 2]
b_ch1_plot_freq = b_ch1_freq[:len(b_ch1) // 2]

plt.plot(b_ch1_plot_freq, np.abs(b_ch1_plot_fft))  # Plot magnitude spectrum
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title('Channel 1 WITH Filter')
plt.show()

# PSD calculation after transform and filters
magnitude_spectrum = np.abs(b_ch1_fft[:len(b_ch1) // 2])
psd = (magnitude_spectrum**2) / (len(b_ch1) * 256)

# Plot PSD
plt.plot(b_ch1_freq[:len(b_ch1) // 2], psd)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power Spectral Density (Î¼VÂ²/Hz)")
plt.title("Channel 1 Power Spectral Density")
plt.show()

# binning function
def computeBands(freq_array, filtered_data):
    bandTotals = [0, 0, 0, 0, 0]
    bandCounts = [0, 0, 0, 0, 0]

    for point in range(len(freq_array)):
        magnitude = np.abs(filtered_data[point])
        if freq_array[point] < 4 and magnitude > 0:
            bandTotals[0] += magnitude
            bandCounts[0] += 1
        elif freq_array[point] < 8 and magnitude > 0:
            bandTotals[1] += magnitude
            bandCounts[1] += 1
        elif freq_array[point] < 12 and magnitude > 0:
            bandTotals[2] += magnitude
            bandCounts[2] += 1
        elif freq_array[point] < 30 and magnitude > 0:
            bandTotals[3] += magnitude
            bandCounts[3] += 1
        elif freq_array[point] < 100 and magnitude > 0:
            bandTotals[4] += magnitude
            bandCounts[4] += 1

    # Calculate the average of all points
    bandAverages = np.array(bandTotals) / np.array(bandCounts)
    return bandAverages

# plot bins
def plotBands(bands):
    binNames = ["Delta", "Theta", "Alpha", "Beta", "Gamma"]
    plt.ylabel("Amplitude")
    plt.bar(binNames, bands, color="#7967e1")
    plt.title('Bored Band Distribution')
    plt.show()
    plt.clf()


# Step 3: Apply the Inverse Fourier Transform
# ============================================
# Create a copy of the Fourier-transformed data
b_ch1_fft_copy = np.copy(b_ch1_fft)

# Apply the Inverse Fourier Transform to the copy
b_ch1_inverse_fft = np.abs(np.fft.ifft(b_ch1_fft_copy))

time = np.arange(0, 256, 1)

plt.figure(figsize=(16, 10))
print("Filtered Plot")
plt.plot(time, b_ch1_inverse_fft[:256])
plt.xlabel("Sample")
plt.ylabel("Î¼V")
plt.title('Channel 1 Inverse after filter')
plt.show()
plt.clf()


# In[8]:


# CHANNEL 2
# ============================================

import numpy as np
import matplotlib.pyplot as plt

# Assuming b_ch1_ contains your channel data
b_ch2_fft = np.fft.fft(b_ch2)
b_ch2_freq = np.fft.fftfreq(len(b_ch2)) * 256

# Compute magnitude spectrum
magnitude_spectrum = np.abs(b_ch2_fft[:len(b_ch2) // 2])

# Plot the magnitude spectrum
plt.plot(b_ch2_freq[:len(b_ch2) // 2], magnitude_spectrum)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title('Channel 2 No Filter')
plt.show()

# Assuming b_ch1_ contains your channel data
cut_freq = 60
tolerance = 2

# Apply Fourier Transform
b_ch2_fft = np.fft.fft(b_ch2)
b_ch2_freq = np.fft.fftfreq(len(b_ch2)) * 256

# Define a threshold frequency
threshold_freq = 100

# Remove frequencies greater than the threshold
b_ch2_fft[np.abs(b_ch2_freq) > threshold_freq] = 0

# Remove powerline interference
index_to_remove = np.where((b_ch2_freq >= cut_freq - tolerance) & (b_ch2_freq <= cut_freq + tolerance))
b_ch2_fft[index_to_remove] = 0

# Plot the modified result
b_ch2_plot_fft = b_ch2_fft[:len(b_ch2) // 2]
b_ch2_plot_freq = b_ch2_freq[:len(b_ch2) // 2]

plt.plot(b_ch2_plot_freq, np.abs(b_ch2_plot_fft))  # Plot magnitude spectrum
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title(' Channel 2 WITH Filter')
plt.show()

# PSD calculation after transform and filters
magnitude_spectrum = np.abs(b_ch2_fft[:len(b_ch2) // 2])
psd = (magnitude_spectrum**2) / (len(b_ch2) * 256)

# Plot PSD
plt.plot(b_ch2_freq[:len(b_ch2) // 2], psd)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power Spectral Density (Î¼VÂ²/Hz)")
plt.title("Channel 2 Power Spectral Density")
plt.show()



# Step 3: Apply the Inverse Fourier Transform
# ============================================
# Create a copy of the Fourier-transformed data
b_ch2_fft_copy = np.copy(b_ch2_fft)

# Apply the Inverse Fourier Transform to the copy
b_ch2_inverse_fft = np.abs(np.fft.ifft(b_ch2_fft_copy))

time = np.arange(0, 256, 1)

plt.figure(figsize=(16, 10))
print("Filtered Plot")
plt.plot(time, b_ch2_inverse_fft[:256])
plt.xlabel("Sample")
plt.ylabel("Î¼V")
plt.title('Channel 2 Inverse after filter')
plt.show()
plt.clf()


# In[5]:


# CHANNEL 3
# ============================================

import numpy as np
import matplotlib.pyplot as plt

# Assuming b_ch1_ contains your channel data
b_ch3_fft = np.fft.fft(b_ch3)
b_ch3_freq = np.fft.fftfreq(len(b_ch3)) * 256

# Compute magnitude spectrum
magnitude_spectrum = np.abs(b_ch3_fft[:len(b_ch3) // 2])

# Plot the magnitude spectrum
plt.plot(b_ch3_freq[:len(b_ch3) // 2], magnitude_spectrum)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title('Channel 3 No Filter')
plt.show()

# Assuming b_ch1_ contains your channel data
cut_freq = 60
tolerance = 2

# Apply Fourier Transform
b_ch3_fft = np.fft.fft(b_ch3)
b_ch3_freq = np.fft.fftfreq(len(b_ch3)) * 256

# Define a threshold frequency
threshold_freq = 100

# Remove frequencies greater than the threshold
b_ch3_fft[np.abs(b_ch3_freq) > threshold_freq] = 0

# Remove powerline interference
index_to_remove = np.where((b_ch3_freq >= cut_freq - tolerance) & (b_ch3_freq <= cut_freq + tolerance))
b_ch3_fft[index_to_remove] = 0

# Plot the modified result
b_ch3_plot_fft = b_ch3_fft[:len(b_ch3) // 2]
b_ch3_plot_freq = b_ch3_freq[:len(b_ch3) // 2]

plt.plot(b_ch3_plot_freq, np.abs(b_ch3_plot_fft))  # Plot magnitude spectrum
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title('Channel 3 WITH Filter')
plt.show()

# PSD calculation after transform and filters
magnitude_spectrum = np.abs(b_ch3_fft[:len(b_ch3) // 2])
psd = (magnitude_spectrum**2) / (len(b_ch3) * 256)

# Plot PSD
plt.plot(b_ch3_freq[:len(b_ch3) // 2], psd)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power Spectral Density (Î¼VÂ²/Hz)")
plt.title("Channel 3 Power Spectral Density")
plt.show()

# Step 3: Apply the Inverse Fourier Transform
# ============================================
# Create a copy of the Fourier-transformed data
b_ch3_fft_copy = np.copy(b_ch3_fft)

# Apply the Inverse Fourier Transform to the copy
b_ch3_inverse_fft = np.abs(np.fft.ifft(b_ch3_fft_copy))

time = np.arange(0, 256, 1)

plt.figure(figsize=(16, 10))
print("Filtered Plot")
plt.plot(time, b_ch3_inverse_fft[:256])
plt.xlabel("Sample")
plt.ylabel("Î¼V")
plt.title('Channel 3 Inverse after filter')
plt.show()
plt.clf()


# In[6]:


# CHANNEL 4
# ============================================

import numpy as np
import matplotlib.pyplot as plt

# Assuming b_ch1_ contains your channel data
b_ch4_fft = np.fft.fft(b_ch4)
b_ch4_freq = np.fft.fftfreq(len(b_ch4)) * 256

# Compute magnitude spectrum
magnitude_spectrum = np.abs(b_ch4_fft[:len(b_ch4) // 2])

# Plot the magnitude spectrum
plt.plot(b_ch4_freq[:len(b_ch4) // 2], magnitude_spectrum)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title('Channel 4 No Filter')
plt.show()

# Assuming b_ch1_ contains your channel data
cut_freq = 60
tolerance = 2

# Apply Fourier Transform
b_ch4_fft = np.fft.fft(b_ch4)
b_ch4_freq = np.fft.fftfreq(len(b_ch4)) * 256

# Define a threshold frequency
threshold_freq = 100

# Remove frequencies greater than the threshold
b_ch4_fft[np.abs(b_ch4_freq) > threshold_freq] = 0

# Remove powerline interference
index_to_remove = np.where((b_ch4_freq >= cut_freq - tolerance) & (b_ch4_freq <= cut_freq + tolerance))
b_ch4_fft[index_to_remove] = 0

print("confirming we are saving filtered fft data", b_ch4_fft) #-------------------------------------------------------------------------------------------------------------------------

# Plot the modified result
b_ch4_plot_fft = b_ch4_fft[:len(b_ch4) // 2]
b_ch4_plot_freq = b_ch4_freq[:len(b_ch4) // 2]

plt.plot(b_ch4_plot_freq, np.abs(b_ch4_plot_fft))  # Plot magnitude spectrum
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title('Channel 4 WITH Filter')
plt.show()

# PSD calculation after transform and filters
magnitude_spectrum = np.abs(b_ch4_fft[:len(b_ch4) // 2])
psd = (magnitude_spectrum**2) / (len(b_ch4) * 256)

# Plot PSD
plt.plot(b_ch4_freq[:len(b_ch4) // 2], psd)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power Spectral Density (Î¼VÂ²/Hz)")
plt.title("Channel 4 Power Spectral Density")
plt.show()

# Step 3: Apply the Inverse Fourier Transform
# ============================================
# Create a copy of the Fourier-transformed data
b_ch4_fft_copy = np.copy(b_ch4_fft)

# Apply the Inverse Fourier Transform to the copy
b_ch4_inverse_fft = np.abs(np.fft.ifft(b_ch4_fft_copy))

time = np.arange(0, 256, 1)

plt.figure(figsize=(16, 10))
print("Filtered Plot")
plt.plot(time, b_ch4_inverse_fft[:256])
plt.xlabel("Sample")
plt.ylabel("Î¼V")
plt.title('Channel 4 Inverse after filter')
plt.show()
plt.clf()


# In[9]:


# Channel 1
b_ch1_bands_before_ifft = computeBands(b_ch1_freq, b_ch1_fft)

# Plot bands before IFFT
plotBands(b_ch1_bands_before_ifft)

# Channel 2
b_ch2_bands_before_ifft = computeBands(b_ch2_freq, b_ch2_fft)

# Plot bands before IFFT
plotBands(b_ch2_bands_before_ifft)

# Channel 3
b_ch3_bands_before_ifft = computeBands(b_ch3_freq, b_ch3_fft)

# Plot bands before IFFT
plotBands(b_ch3_bands_before_ifft)

# Channel 4
b_ch4_bands_before_ifft = computeBands(b_ch4_freq, b_ch4_fft)

# Plot bands before IFFT
plotBands(b_ch4_bands_before_ifft)

```
